```python
new=df.copy()
new
```

```python

conditions = [
    new['How many hours do you spend on social media in the evening?'] == 'Less than 1 hour',
    new['How many hours do you spend on social media in the evening?'] == '1–2 hours',
    new['How many hours do you spend on social media in the evening?'].isin(['3–4 hours', '4-5 hours']),
    new['How many hours do you spend on social media in the evening?'] == '5+ hours'
]
choices = ['low', 'moderate', 'high', 'very high']

# Apply np.select
new['Usage_in_evening'] = np.select(conditions, choices, default='unknown')

new
```

```python
new.drop('How would you rate your mood in the morning (on a scale of 1–10)?',inplace=True,axis=1)
```

```python
new['How would you rate your mood in the evening (on a scale of 1–10)?'].value_counts()
```

```python
new['How many hours of sleep do you get on average per night?'].value_counts()

```

```python
new['Do you feel stressed during the day?'].value_counts()
```

```python
# Assign scores to the rest of the columns
mood_change_scores = {'Positive change': 2, 'Negative change': -2, 'No change': 0}
sleep_scores = {'5–6 hours': 1, '9+ hours': 2, 'Less than 5 hours': -1,'7–8 hours':3}
stress_scores = {'No': 3,"No, I don't feel stressed.":3, 'Yes': -2, 'Studying': -1, 'Yes, and it significantly impacts my day.': -3,
                  'Yes, but only in specific situations (e.g., work, studies).':-1,'Yes, but it is manageable.':2}

new['Mood_Score'] =new['How would you rate your mood in the evening (on a scale of 1–10)?']
new['Mood_Change_Score'] = new['Do you notice a change in your mood after using social media in the evening?'].map(mood_change_scores)
new['Sleep_Score'] = new['How many hours of sleep do you get on average per night?'].map(sleep_scores)
new['Stress_Score'] = new['Do you feel stressed during the day?'].map(stress_scores)
new['addiction_score']=new['Do you think your social media usage has become a habit or addiction?'].map({'Yes':-1,'No':1})

# 3. Create the new feature by summing the scores
new['Aggregated_Score'] = new[['Mood_Score', 'Mood_Change_Score', 'Sleep_Score', 'Stress_Score','addiction_score']].sum(axis=1)

# View the updated DataFrame
new
```

```python
final_df=new[['Usage_in_evening','Mood_Score','Mood_Change_Score',
              'Sleep_Score','Stress_Score','addiction_score','Aggregated_Score']]
```

```python
final_df=final_df[['Usage_in_evening','Aggregated_Score']]
final_df
```

```python
final_df['Aggregated_Score'].describe()
```

```python


# Define score bins
bins = [0, 4, 12, float('inf')]  # Define bin ranges
labels = ['low', 'moderate', 'high']  # Define labels for the bins

# Categorize Aggregated_Score
final_df['Score_Category'] = pd.cut(final_df['Aggregated_Score'], bins=bins, labels=labels)

print(final_df)
```

```python
# Create contingency table
contingency_table = pd.crosstab(final_df['Usage_in_evening'], final_df['Score_Category'])
print(contingency_table)
```

```python
from scipy.stats import chi2_contingency

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-Square Statistic:", chi2)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:", expected)
```
